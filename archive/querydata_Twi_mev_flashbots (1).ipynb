```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import snscrape.modules.twitter
```

```python
# Creating list to append tweet data to
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to lifst
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#MEV since:2019-01-01 until:2022-10-01','corona lang:en').get_items()):
    tweets_list.append([tweet.date,  tweet.content]) #tweet.id
    print(tweet.date)
```

```python
result= pd.DataFrame(tweets_list, columns=['Date', 'Tweets'])
result

```

```python
result.to_csv('mev.csv')
```
